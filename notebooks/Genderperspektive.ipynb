## Prominente Autorinnen in CK

```python
f = ck.get_authors(gender="female", threshold=5)
```

```python
m = ck.get_authors(gender="male", threshold=5)
```

```python
# print top 15 female authors with the most publications
f.df[["name","article_count"]].sort_values(["article_count"], ascending=False)[:10]
```

Ein großteil der in CK häufig publizierenden Frauen sind primär Schriftstellerinnen (Tawara Michi, Tanabe Seiko, Nogami Yaeko, Sono Ayako, Kouno Taeko, Enchi Fumiko, Ariyoshi Sawako, Hiraiwa Yumie, Miyamoto Yuriko). Ausnahmen sind die Kimono Designerin Uno Chiyo, die Historikerin Shiono Nanami, die TV-Journalistin Sakurai Yoshiko.

Im Vergleich dazu ist das berufliche Sprektrum der männlichen Autoren größer: Akademiker (Yoshino Sakuzou), Historiker (Nanabe Teruhiko), Schriftsteller (Masamune Hakuchou, Nawa Fumio, Ryoutarou Shiba), Spieler (Takemiya Masaki, Yonenaga Kunio), Ökonom (Eikan Kyuu), Dichter (Okano Hirohiko), Journalisten (Baba Tsunego)
