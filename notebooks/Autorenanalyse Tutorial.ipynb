## Laden der Zeitschriftendaten und Generation von Autorenlisten
Um die Magazindaten zu laden, instanziere die Klasse `Magazine`.

```python
#load chuo koron magazine
ck = Magazine()
```

`Magazine` hat die Funktion `get_authors`, welche eine neue Au* torenliste (Klasse `AuthorList`) generiert.
`get_authors` hat drei Parameter:
* `year_from`
* `year_to`
* `threshold`: Mindestanzahl von Artikel eines Authors im angegebenen Zeitraum

```python
#load all authors
authors = ck.get_authors(1920, 2000)
```

```python
#load authors between 1920 and 1930 with min 5 articles
authors2 = ck.get_authors(1920, 1930, threshold=5)
```

## Verfügbare Daten

Um einen allgemeinen Überblick über die verfügbaren Daten in der jeweiligen Autorenliste zu bekommen, kann man die Funktion `available_data` ausführen.

```python
#plot available data
authors.available_data()
```

Allgemeines zur Datenlagen: 
* Fast alle Autorendatensätze beinhalten ein Geburtsjahr
* Angaben zu Bildung und Geburtsort gibt es für gut die Hälte der Datensätze
* Genderangabe für gut zwei Drittel

```python
authors2 = ck.get_authors(1920, 2000, threshold=10)
```

```python
authors2.available_data()
```

Erstellt man ein Autorendatenset mit einem höheren Schwellwert (z.B. min 10 veröffentlichte Artikel), verbessert sich die Datenlage deutlich.

## Einfache Statistiken

Die Klasse `AuthorList` verfügt auch über eine Reihe von Funktionen um einfache Statistiken (+ Visualisierungen) zu erstellen: 
* `gender_stats`
* `education_stats`
* `birth_year_stats`

### Geschlechterverhältnis

```python
authors.gender_stats()
```

### Universitäre Bildung

```python
a1 = ck.get_authors(1945,1960)
a1.education_stats()
```

```python
a2 = ck.get_authors(1975,1990)
a2.education_stats()
```

### Geburtsjahr 

```python
a = ck.get_authors(1960, 1970, threshold=5)
```

```python
a.birth_year_stats()
```

### Todesjahre

```python
a = ck.get_authors(1920, 1940, threshold=5)
```

```python
a.death_year_stats()
```

```python
a.death_year_stats(cumsum=True)
```

## Karte der Geburtsorte

Mit der Funktion `birth_map` einer Autorenliste kann man auch eine Karte der Geburtsorte der Autoren erstellen.
Die `birth_map` Funktion hat den Parameter `marker`. Ist dieser `True`, werden Marker gesetzt, wenn `False` wird die Karte als Heatmap angezeigt.

```python
a = ck.get_authors(1950, 1960, threshold=5)
```

```python
a.birth_map(marker=True)
```
